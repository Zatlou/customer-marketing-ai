```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Chargement du dataset clustérisé
df = pd.read_csv("../data/processed/marketing_campaign_clustered.csv")

# 2. Visualisation de la cible 'Response'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Response')
plt.title("Répartition des réponses à la campagne")
plt.show()

# 3. Validation des clusters (Revenus vs Dépenses)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='TotalSpending', hue='ClusterID', palette='viridis')
plt.title("Segments Clients : Revenus vs Dépenses")
plt.show()

# 4. Taux de réponse par cluster
df.groupby('ClusterID')['Response'].mean().plot(kind='bar')
plt.title("Taux de réponse par Segment")
plt.ylabel("Proportion de Response=1")
plt.show()
```
